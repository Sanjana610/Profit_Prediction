```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("advertising.csv")
```

```python
df.head()
```

```python
plt.xlabel('TV')
plt.ylabel('Sales')
plt.scatter(df.TV,df.Sales, color='red',marker = '+')
```

```python
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.scatter(df.Radio,df.Sales, color='blue',marker = '*')
```

```python
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.scatter(df.Newspaper,df.Sales, color='green',marker = '^')
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['TV','Radio','Newspaper']],df.Sales,test_size = 0.2)
```

```python
len(X_train)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
